** Zadatak ** U stupac d.odst upisati odstupanje od prosjeka.

```python
prosjeci = df.groupby("mon").aggregate(np.mean).avg.values
prosjeci[0]
def f_odst(m, a):
    return (a - prosjeci[m-1])
```

```python
df.odst = f_odst(m = df.mon, a = df.avg)
df.head()
```
